```python
# train_door_model.py
from ultralytics import YOLO
from pathlib import Path
```

```python

# Path to your merged dataset's data.yaml
DATA_YAML = "/content/drive/MyDrive/dataset_door/1/data.yaml"  # replace if different

# Choose YOLOv8 model version
# Options: yolov8n.pt (nano), yolov8s.pt (small), yolov8m.pt (medium), yolov8l.pt (large), yolov8x.pt (extra large)
MODEL = "yolov8n.pt"
```

```python
# Training hyperparameters
EPOCHS = 50
BATCH_SIZE = 16
IMG_SIZE = 640
EXPERIMENT_NAME = "door_openclose_exp"

# Initialize YOLO model
model = YOLO(MODEL)
```

```python


# Start training
model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    batch=BATCH_SIZE,
    imgsz=IMG_SIZE,
    name=EXPERIMENT_NAME,
    device=0  # use '0' for GPU, 'cpu' for CPU
)

# After training, the best.pt weights will be saved in ./runs/detect/door_openclose_exp/weights/best.pt
print("Training complete. Check ./runs/detect/door_openclose_exp/weights/best.pt")
```
